```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

## Determine the hour by creating list from TIME OCC strings to find highest frequency

```python
crime_hour = []
for hour in crimes["TIME OCC"]:
    crime_hour.append(hour[:2])
```

```python
# Add created list to crimes dataframe as a Series
crimes["crime_hour"] = pd.Series(crime_hour)
```

```python
# View the data to validate result
sns.countplot(data=crimes, x="crime_hour")
```

```python
# Convert crime_hour to int and save as requested variable
peak_crime_hour = int(crimes.crime_hour.mode())
```

## Find areas with crimes in night crime hours then count

```python
night_crime_hours = ["22","23","0","1","2","3"]
```

```python
night_crimes = crimes[crimes.crime_hour.isin(night_crime_hours)]
```

```python
peak_night_crime_location = night_crimes["AREA NAME"].value_counts().index[0]
```

## Create bins for various ages and then groupby these ages

```python
labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]
bins = [0,17,25,34,44,54,64, 120]
```

```python
crimes["victim_age_groups"] = pd.cut(crimes["Vict Age"], labels = labels, bins = bins)
```

```python
victim_ages = crimes["victim_age_groups"].value_counts()
```
